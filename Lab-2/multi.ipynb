```python
import numpy as np
import matplotlib.pyplot as plt
strn = input()
arr = np.array([int(i) for i in strn])
print(arr)
```

## Original

```python
plt.figure(figsize=(12, 6), dpi=100, facecolor='whitesmoke')
plt.title("Original Signal", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(arr)+1), np.append(arr, arr[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Encoding

```python
encoded = []
cur = [-1, 0, 1, 0]
j = 1

for i in arr:
    if i == 0:
        encoded.append(cur[j])
    elif i == 1:
        j = (j+1)%4
        encoded.append(cur[j])
        
print(encoded)
```

```python
plt.figure(figsize=(12, 6), dpi=100, facecolor='whitesmoke')
plt.title("Multi-Level Line Encoding", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(encoded)+1), np.append(encoded, encoded[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Decoding

```python
decode = []
cur = [-1,0,1,0]
j = 1

for i in encoded:
    if i == cur[j]:
        decode.append(0)
    else:
        j = (j+1)%4
        decode.append(1)
print(decode)
```

```python
plt.figure(figsize=(12, 6), dpi=100, facecolor='whitesmoke')
plt.title("Multi-Level Line Decoding", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(decode)+1), np.append(decode, decode[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

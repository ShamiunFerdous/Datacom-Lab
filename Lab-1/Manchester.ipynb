```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
strn = '10111001'
arr = np.array([int(i) for i in strn])
print("Original bits : ", arr)
```

## Graphical representation of input signal

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Original Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(arr)+1), np.append(arr, arr[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```

## Logic for encoding

```python
encoded = []

for i in arr:
    if i == 1:
        encoded.append(1)
        encoded.append(0)
    else:
        encoded.append(0)
        encoded.append(1)
print(encoded)
```

## Graphical Representation of encoded signal

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Manchester Encoded Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(encoded)+1), np.append(encoded, encoded[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```

## Logic for Decoding

```python
decode = []

for i in range(0,len(encoded),2):
    if encoded[i] == 1:
        decode.append(1)
    else:
        decode.append(0)

strn = "".join(str(decode))
print(strn)
```

## Graphical representaion of decoded signal

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Manchester Decoded Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(decode)+1), np.append(decode, decode[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```

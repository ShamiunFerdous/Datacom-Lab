```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
strn = input()
arr = np.array([int(i) for i in strn])
print("Original bits : ", arr)
```

```python
plt.figure(figsize=(10,5), dpi=100, facecolor='whitesmoke')
plt.title("Original Signal", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(arr)+1), np.append(arr, arr[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Logic for encoding

```python
encoded = []

for i in arr:
    if i == 1:
        encoded.append(1)
        encoded.append(0)
    else:
        encoded.append(0)
        encoded.append(1)
print(encoded)
```

## Graphical Representation of encoded signal

```python
plt.figure(figsize=(12,6), dpi=100, facecolor='whitesmoke')
plt.title("Manchester Encoded Signal", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(encoded)+1), np.append(encoded, encoded[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Logic for Decoding

```python
decode = []

for i in range(0,len(encoded),2):
    if encoded[i] == 1:
        decode.append(1)
    else:
        decode.append(0)

strn = "".join(str(decode))
print(strn)
```

## Graphical representaion of decoded signal

```python
plt.figure(figsize=(10,5), dpi=100, facecolor='whitesmoke')
plt.title("Manchester Decoded Signal", fontsize=22, fontweight='bold', pad=20)
plt.step(range(len(decode)+1), np.append(decode, decode[-1]), where='post', 
         color='#1f77b4', linewidth=3, linestyle='-', marker='o', markersize=5, markerfacecolor='red')
plt.yticks([0, 1], ['Low (0)', 'High (1)'], fontsize=14)
plt.xlabel('Time', fontsize=16, fontweight='bold', labelpad=15)
plt.ylabel('Voltage (V)', fontsize=16, fontweight='bold', labelpad=15)
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)
plt.minorticks_on()
plt.grid(True, which='minor', axis='both', linestyle=':', color='lightgray', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

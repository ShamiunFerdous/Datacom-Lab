```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
strn = '10111001'
arr = np.array([int(i) for i in strn])
print("Original bits : ", arr) 
```

## Graphical representation of the original input

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("Original Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(arr)+1), np.append(arr, arr[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```

## Logic for encoding

```python
encoded = []
for i in arr:
    if i == 1:
        encoded.append(1)
    else:
        encoded.append(0)
print(encoded)
```

## Graphical representation of encoded signal

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Encoded Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(encoded)+1), np.append(encoded, encoded[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```

## Logic for Decoding

```python
decode = []
for i in encoded:
    if i == 1:
        decode.append(1)
    else:
        decode.append(0)
strn = "".join(str(i) for i in decode)
print(strn)
```

## Graphical representation of Decoded signal

```python
plt.figure(figsize=(10, 5), dpi=200, facecolor='lightgray')
plt.title("NRZ-L Decoded Signal", fontdict={'fontname': 'Kalpurush', 'fontsize': 20})
plt.step(range(len(encoded)+1), np.append(decode, decode[-1]), where='post', color='blue', linewidth=2)
plt.yticks([0, 1], ['Low', 'High'])
plt.xlabel('Time', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.ylabel('Voltage', fontdict={'fontname': 'Kalpurush', 'fontsize': 15})
plt.grid(True)
plt.show()
```
